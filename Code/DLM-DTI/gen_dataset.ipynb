```python
import pandas as pd

davis_train = pd.read_csv("data/DAVIS_train.csv")
davis_valid = pd.read_csv("data/DAVIS_valid.csv")
davis_test = pd.read_csv("data/DAVIS_test.csv")

davis_train.loc[:, "Source"] = "DAVIS"
davis_valid.loc[:, "Source"] = "DAVIS"
davis_test.loc[:, "Source"] = "DAVIS"

KIBA_train = pd.read_csv("data/KIBA_train.csv")
KIBA_valid = pd.read_csv("data/KIBA_valid.csv")
KIBA_test = pd.read_csv("data/KIBA_test.csv")

KIBA_train.loc[:, "Source"] = "KIBA"
KIBA_valid.loc[:, "Source"] = "KIBA"
KIBA_test.loc[:, "Source"] = "KIBA"
```

```python
train_df = pd.concat([davis_train, KIBA_train]).reset_index(drop=True)
valid_df = pd.concat([davis_valid, KIBA_valid]).reset_index(drop=True)
test_df = pd.concat([davis_test, KIBA_test]).reset_index(drop=True)
```

```python
train_df.to_csv("data/train_dataset.csv", index=False)
valid_df.to_csv("data/valid_dataset.csv", index=False)
test_df.to_csv("data/test_dataset.csv", index=False)
```
